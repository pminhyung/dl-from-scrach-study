```python
def AND_gate(x1, x2):
    w1 = 0.5
    w2 = 0.5
    b = -0.7 #bias 편향; 임계치를 0으로 하기위해 합하여지는 값
    '''본예시에서 임계치(theta)값은 0.7'''
    result = x1*w1 + x2*w2 + b
    
    if result <= 0:
        return 0
    else:
        return 1
```

```python
AND_gate(0, 0), AND_gate(0, 1), AND_gate(1, 0), AND_gate(1, 1)
```

```python
def NAND_gate(x1, x2):
    w1=-0.5
    w2=-0.5
    b=0.7
    result = x1*w1 + x2*w2 + b
    if result <= 0:
        return 0
    else:
        return 1
```

```python
NAND_gate(0, 0), NAND_gate(0, 1), NAND_gate(1, 0), NAND_gate(1, 1)
```

```python
def OR_gate(x1, x2):
    w1= 0.6
    w2= 0.6
    b= -0.5
    result = x1*w1 + x2*w2 + b
    if result <= 0:
        return 0
    else:
        return 1
```

```python
OR_gate(0, 0), OR_gate(0, 1), OR_gate(1, 0), OR_gate(1, 1)
```
